Функция build_convex_hull_binary_image строит 45-выпуклую оболочку для объектов на картинке.

```python
A = (np.array([[1, 1, 1], [0, 0, 1], [0, 0, 0]], dtype=np.bool),
     np.array([[1, 1, 0], [1, 0, 0], [1, 0, 0]], dtype=np.bool),
     np.array([[0, 0, 0], [1, 0, 0], [1, 1, 1]], dtype=np.bool),
     np.array([[0, 0, 1], [0, 0, 1], [0, 1, 1]], dtype=np.bool),
     np.array([[1, 1, 1], [1, 0, 0], [0, 0, 0]], dtype=np.bool),
     np.array([[1, 0, 0], [1, 0, 0], [1, 1, 0]], dtype=np.bool),
     np.array([[0, 0, 0], [0, 0, 1], [1, 1, 1]], dtype=np.bool),
     np.array([[0, 1, 1], [0, 0, 1], [0, 0, 1]], dtype=np.bool))
B = [np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]], dtype=np.bool) for _ in range(8)]
convex_hull = build_convex_hull_binary_image(image, A, B)
plt.imshow(convex_hull, cmap='Greys')
```
